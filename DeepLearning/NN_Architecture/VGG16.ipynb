```python
from keras.models import Sequential
```

```python
vgg16 = Sequential()
```

2 x convolution layer of 64 channel of 3x3 kernal and same padding

→ 1 x maxpool layer of 2x2 pool size and stride 2x2

→ 2 x convolution layer of 128 channel of 3x3 kernal and same padding

→ 1 x maxpool layer of 2x2 pool size and stride 2x2

→ 3 x convolution layer of 256 channel of 3x3 kernal and same padding

→ 1 x maxpool layer of 2x2 pool size and stride 2x2

→ 3 x convolution layer of 512 channel of 3x3 kernal and same padding

→ 1 x maxpool layer of 2x2 pool size and stride 2x2

→ 3 x convolution layer of 512 channel of 3x3 kernal and same padding

→ 1 x maxpool layer of 2x2 pool size and stride 2x2

I also add relu (Rectified Linear Unit) activation to each layer so that all the negative values are not passed to the next layer.

```python
from keras.layers import Dense , Conv2D , MaxPooling2D , Flatten
```

### Model Architecture

```python

vgg16.add(Conv2D(filters=64 , kernel_size = (3,3), input_shape = (224 , 224 ,3) , activation='relu' , padding='same'))

vgg16.add(Conv2D(filters=64 , kernel_size = (3,3) ,activation='relu', padding='same'))
vgg16.add(MaxPooling2D(strides=(2,2), pool_size=(2,2)))

vgg16.add(Conv2D(filters=128 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=128 , kernel_size = (3,3) , padding='same'))
vgg16.add(MaxPooling2D(pool_size = (2,2) , strides=(2,2)))

vgg16.add(Conv2D(filters=256 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=256 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=256 , kernel_size = (3,3) , padding='same'))
vgg16.add(MaxPooling2D(pool_size = (2,2) , strides=(2,2)))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))
vgg16.add(MaxPooling2D(pool_size = (2,2) , strides=(2,2)))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))

vgg16.add(Conv2D(filters=512 , kernel_size = (3,3) , padding='same'))
vgg16.add(MaxPooling2D(pool_size = (2,2) , strides=(2,2)))
```

```python
vgg16.add(Flatten())
```

```python
vgg16.add(Dense(units = 4096 ,activation='relu'))

vgg16.add(Dense(units = 4096 ,activation='relu'))

vgg16.add(Dense(units = 1000 ,activation='softmax'))
```
