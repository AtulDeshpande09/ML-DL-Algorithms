```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Conv2D , MaxPooling2D , Flatten , Dense , Dropout
```

### Model Architecture

```python
# Layer 1
model.add(Conv2D(96 , strides =4 , kernel_size = (11,11),input_shape = (227,227,3) , activation = 'relu'))
model.add(MaxPooling2D(pool_size = (3,3), strides = (2,2)))

# Layer 2
model.add(Conv2D(256 , strides =1 , kernel_size = (5,5), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (3,3), strides = (2,2)))

# Layer 3
model.add(Conv2D(384 , strides =1 , kernel_size = (3,3), activation = 'relu',padding='same'))

# Layer 4
model.add(Conv2D(384 , strides =1 , kernel_size = (3,3), activation = 'relu',padding='same'))

# Layer 5
model.add(Conv2D(256 , strides =1 , kernel_size = (3,3), activation = 'relu',padding='same'))
model.add(MaxPooling2D(pool_size = (3,3), strides = (2,2)))


model.add(Flatten())

# Layer 6
model.add(Dense(9216,activation='relu'))

# Layer 7
model.add(Dense(4096,activation='relu'))

# Layer 8
model.add(Dense(4096,activation='relu'))

# Output Layer
model.add(Dense(1000,activation='softmax'))
```
