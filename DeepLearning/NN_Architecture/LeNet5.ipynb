```python
from keras.models import Sequential
```

```python
lenet5 = Sequential()
```

```python
from keras.layers import Conv2D, AveragePooling2D , Flatten , Dense
```

```python
lenet5.add(Conv2D(6 ,(5,5), padding='same', input_shape = (28,28,1),activation='tanh'))
lenet5.add(AveragePooling2D((2,2), strides = (2,2)))

lenet5.add(Conv2D(16 , (5,5), padding ='same' , activation='tanh' , strides=(2,2)))
lenet5.add(AveragePooling2D((2,2), strides = (2,2)))
lenet5.add(Flatten())

lenet5.add(Dense(120 , activation='tanh'))
lenet5.add(Dense(84 , activation='tanh'))

lenet5.add(Dense(10 , activation='softmax'))
```
