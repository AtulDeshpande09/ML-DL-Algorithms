```python
from robust_scaling import RobustScaler
```

```python
scaler = RobustScaler()

data = [40,30,45,20,45,33,21,25,55,56,66]

x_scaled = scaler.fit_transform(data)
x_scaled
```

#### Reverting Data

```python
x_original = scaler.inverse_transform(x_scaled)
x_original
```

#### Visualization

```python
import matplotlib.pyplot as plt
```

```python
# Plot histograms for original and scaled data
plt.figure(figsize=(6,4))

plt.subplot(1, 2, 1)
plt.hist(data, bins=10, edgecolor='black')
plt.title("Original Data Distribution")

plt.subplot(1, 2, 2)
plt.hist(x_scaled, bins=10, edgecolor='black')
plt.title("Scaled Data Distribution")

plt.tight_layout()
plt.show()
```
