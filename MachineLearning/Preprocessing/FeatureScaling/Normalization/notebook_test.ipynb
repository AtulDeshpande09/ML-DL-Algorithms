```python
from Normalize import Normalizer
```

```python
scaler = Normalizer()

data = [40,30,45,20,45,33,21,25,55,56,66]

x_scaled = scaler.fit_transform(data)
x_scaled
```

#### Reverting Data

```python
x_original = scaler.inverse_transform(x_scaled)
x_original
```

#### Visualization

```python
x1 = [211,322,443,501,300,453,321,453]
x1_scaled = scaler.fit_transform(x1)
```

```python
x2 = [6564,3455,9874,4563,7644,4987,2222,2334]
x2_scaled = scaler.fit_transform(x2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original Data")
sns.scatterplot(x=x1,y=x2)
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.subplot(1,2,2)
plt.title("Scaled Data")
sns.scatterplot(x=x1_scaled,y=x2_scaled)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
