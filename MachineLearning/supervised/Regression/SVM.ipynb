### Regression Dataset
We will create dataset using `make_regression` .

Dataset will have :
- 1000 samples (`n_samples`) ( number of rows )
- 2 features (`n_features` ) ( number of feature columns )

```python
from sklearn.datasets import make_regression
```

```python
X , y = make_regression(n_samples=1000, n_features=2, random_state=42, bias = 5, noise = 1)
```

```python
X.shape
```

```python
y.shape
```

### Splitting Dataset

Let's split the dataset using `train_test_split` .
- Training Data = 75%
- Testing Data = 25%

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state = 42 , test_size=0.25 )
```

```python
X_train.shape
```

```python
y_train.shape
```

### Model Training

We will train a SVM model for regression using `SVR` from `svm` .

We will use `linear` kernel ( you can also try with other kernels such as `rbf` , `poly` , `sigmoid` )

```python
from sklearn.svm import SVR
```

```python
model = SVR(kernel='linear')
```

```python
model.fit(X_train,y_train)
```

### Prediction , Evaluation

We will predict the values of y using `model.predict` .

We will evaluate based on these metrics :
- `Mean Squared Error`
- `Mean Absolute Error`
- `R2 Score`

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
```

```python
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)
r2 = r2_score(y_test , y_pred)
```

```python
print(f"Mean Square Error : {mse:.4f}")
print(f"Mean Absolute Error : {mae:.4f}")
print(f"R2 score : {r2:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```
